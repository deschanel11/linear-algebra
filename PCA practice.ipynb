```python
from sklearn.datasets import load_iris

iris = load_iris()

N = 10 # 앞의 10송이만 선택
X = iris.data[:N, :2] #꽃받침 길이와 꽃받침 폭만 선택

#꽃받침 길이, 꽃받침 폭!
X
```

```python
from sklearn.decomposition import PCA

pca1 = PCA(n_components = 1)
X_low = pca1.fit_transform(X)

X_low #1차원 데이터가 나옴. 크기값들.
```

```python
#원점이 아니라 다른 점을 중심으로 길이를 쟀기 때문에 마이너스도 있음.
```

```python
#저 1차원 결과값은 2차원 상에서 어디인지 잘 모르겠음.
#따라서 inverse transform을 해 줌.
#저차원 결과 데이터를 다시 원래의 차원으로 표현해주는 것.

X2 = pca1.inverse_transform(X_low)

X2
```

```python
#import pandas as pd
#import matplotlib.pylab as plt
#import seaborn as sns

plt.figure(figsize=(7, 7))
ax = sns.scatterplot(0, 1, data=pd.DataFrame(X), s=100, color=".2", marker="s")

for i in range(N):
    d = 0.03 if X[i, 1] > X2[i, 1] else -0.04
    ax.text(X[i, 0] - 0.065, X[i, 1] + d, "표본 {}".format(i+1))
    plt.plot([X[i, 0], X2[i, 0]], [X[i, 1], X2[i, 1]], "k--")

plt.plot(X2[:, 0], X2[:, 1], "o-", markersize=10)
plt.plot(X[:, 0].mean(), X[:, 1].mean(), markersize=10, marker="D")
plt.axvline(X[:, 0].mean(), c='r')
plt.ashline(X[:, 1].mean(), c='r')
plt.grid(False)
plt.xlabel("꽃받침의 길이")
plt.ylabel("꽃받침 폭")
plt.title("Iris 데이터의 1차원 차원축소")
plt.axis("equal")
plt.show()
```

```python
#PCA : 잠재변수를 찾는 것이자 더 낮은 차원으로 차원을 축소시키는 행위.
```

```python
#데이터의 평균값 : mean_ 속성으로 보기 가능.

pca1.mean_
```

```python
#주성분벡터(가장 근사하는 데이터를 만드는 단위 기저벡터) -> components_ 속성을 이용해 구할 수 있음.

pca1.components_
```

```python
from sklearn.datasets import fetch_olivetti_faces

faces_all = fetch_olivetti_faces()

K = 20 #20번 인물의 사진만 선택
faces = faces_all.images[faces_all.target == K]
```

```python
faces.shape
```

```python
#정확히는 64*64, 즉 4096차원 데이터임.
64*64
```

```python
#4096차원에 10개의 데이터(x1부터 x10까지 10개의 벡터)가 있는 것.
```

```python
N = 2
M = 5
fig = plt.figure(figsize = (10, 5))
plt.subplots_adjust(top=1, bottom = 0, hspace = 0, wspace = 0.05)
for i in range(N):
    for j in range(M):
        k = i*M+j
        ax = fig.add_subplot(N, M, k+1)
        ax.imshow(faces[k], cmap = plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
plt.suptitle("올리베티 얼굴 사진")
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
pca3 = PCA(n_components = 2)
X3 = faces_all.data[faces_all.target == K]
W3 = pca3.fit_transform(X3)
X32 = pca3.inverse_transform(W3)
```

```python
N = 2
M = 5
fig = plt.figure(figsize = (10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace = 0, wspace = 0.05)
for i in range(N):
    for j in range(M):
        k = i*M+j
        ax = fig.add_subplot(N, M, k+1)
        ax.imshow(X32[k].reshape(64, 64), cmap = plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])

plt.suptitle("주성분 분석으로 근사화한 올리베티 얼굴 사진")
plt.tight_layout()
plt.show()
```

```python
face_mean = pca3.mean_.reshape(64, 64)
face_p1 = pca3.components_[0].reshape(64, 64)
face_p2 = pca3.components_[1].reshape(64, 64)

plt.subplot(131)
plt.imshow(face_mean, cmap = plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("평균 얼굴")
plt.subplot(132)
plt.imshow(face_p1, cmap = plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("주성분 1")
plt.subplot(133)
plt.imshow(face_p2, cmap = plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("주성분 2")
plt.show()
```

```python
N = 2
M = 5
fig = plt.figure(figsize = (10, 5))
plt.subplots_adjust(top=1, bottom = 0, hspace = 0, wspace = 0.05)
for i in range(N):
    for j in range(M):
        k = i*M + j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5*(k-5) if k<5 else 1.5*(k-4)
        ax.imshow(face_mean + w*face_p1, cmap = plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title("주성분1의 비중 = {}".format(w))
plt.suptitle("평균 얼굴에 주성분1을 더한 사진")
plt.tight_layout()
plt.show()
```

```python
N = 2
M = 5
fig = plt.figure(figsize = (10, 5))
plt.subplots_adjust(top = 1, bottom = 0, hspace = 0, wspace = 0.05)
for i in range(N):
    for j in range(M):
        k = i*M + j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5*(k-5) if k<5 else 1.5*(k-4)
        ax.imshow(face_mean + w*face_p2, cmap = plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title("주성분 2의 비중 = {:.1f}".format(w))
plt.suptitle("평균 얼굴에 주성분2를 더한 사진")
plt.tight_layout()
plt.show()
```
